## Cleaning Data

```python
img_predictions_clean = img_predictions.copy()
twitter_arch_clean = twitter_arch.copy()
twitter_json_clean = pd_twitter_json.copy()
```

### Define
We want to keep original ratings so we can drop retweets 
### Code 

```python
twitter_arch_clean = twitter_arch_clean[pd.isnull(twitter_arch_clean.retweeted_status_id)]
twitter_arch_clean = twitter_arch_clean.drop('retweeted_status_id', 1)
twitter_arch_clean = twitter_arch_clean.drop('retweeted_status_user_id', 1)
twitter_arch_clean= twitter_arch_clean.drop('retweeted_status_timestamp' , 1)
```

### Test

```python
twitter_arch_clean.info()
```

### Define

Remove underscore and make 1st letters of dog names as capital in image predictions table

### Code

```python
img_predictions_clean.p1 = img_predictions_clean.p1.str.replace('_', ' ')
img_predictions_clean.p1 = img_predictions_clean.p1.str.title()

img_predictions_clean.p2 = img_predictions_clean.p2.str.replace('_', ' ')
img_predictions_clean.p2 = img_predictions_clean.p2.str.title()

img_predictions_clean.p3 = img_predictions_clean.p3.str.replace('_', ' ')
img_predictions_clean.p3 = img_predictions_clean.p3.str.title()
```

### Test

```python
img_predictions_clean.head()
```

### Define

timestamp in twitter_arch_clean DataFrame should be datetime instead of object(string).

### Code

```python
twitter_arch_clean['timestamp'] = pd.to_datetime(twitter_arch_clean.timestamp)
```

### Test

```python
twitter_arch_clean.info()
```

### Define 

rename columns 

### Code


```python
img_predictions_clean = img_predictions_clean.rename(columns={'p1':'probability1', 'p2':'probability2', 'p3':'probability3'})
```

### Test

```python
img_predictions_clean.sample(5)
```

### Define
Drope columns 'doggo', 'floofer','pupper', 'puppo' and replace with one column 'dog_satge'

### Code

```python
twitter_arch_clean.doggo = twitter_arch_clean.doggo.replace('None', np.nan)
twitter_arch_clean.floofer = twitter_arch_clean.floofer.replace('None', np.nan)
twitter_arch_clean.pupper = twitter_arch_clean.pupper.replace('None', np.nan)
twitter_arch_clean.puppo = twitter_arch_clean.puppo.replace('None', np.nan)
```

```python
twitter_arch_clean['dog_stage'] = twitter_arch_clean[twitter_arch_clean.columns[-4:]].apply(
    lambda x: ','.join(x.dropna().astype(str)),
    axis=1
)
```

```python
twitter_arch_clean = twitter_arch_clean.drop(['doggo','floofer','pupper','puppo'], axis=1)
```

```python
twitter_arch_clean.dog_stage = twitter_arch_clean.dog_stage.replace(r'^\s*$', np.nan, regex=True)
```

### Data Analysis & Visualization

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dfv = pd.read_csv('twitter_archive_master.csv')
```

```python
dfv.info()
```

```python
df_breed = dfv.probability1
```

```python
plt.rcParams['figure.figsize']=(15,15)
df_breed.value_counts(sort=True)[:10].plot.pie(startangle=270, pctdistance=0.8, radius = 0.9)
plt.title('Distribution ofBreeds')
plt.ylabel('')
plt.show();
```

```python
df_breed.head()
```

```python
fav_breed = df_breed.value_counts()
breed_sort = fav_breed.sort_values(ascending = False)
top10_breeds = breed_sort.head(10)
top10_breeds
```

```python
plt.barh(top10_breeds.index, top10_breeds, color = 'Y')
plt.xlabel('Aggregate Favorite Count', fontsize = 14)
plt.title('Top 10 Dog Breeds by Aggregate Favorite Count', fontsize = 16)
plt.gca().invert_yaxis()
plt.show();
```

```python
dfv.groupby('dog_stage')['rating_numerator'].describe()
```

```python
df_name = dfv.name
```

```python
plt.rcParams['figure.figsize']=(15,15)
df_name.value_counts(sort=True)[:10].plot.pie(startangle=270, pctdistance=0.8, radius = 0.9)
plt.title('Distribution ofBreeds')
plt.ylabel('')
plt.show();
```

```python
dfv.groupby('dog_stage')['timestamp'].describe()
```

```python
stage_groups = dfv.groupby('dog_stage')['rating_numerator'].mean()
```

```python
# sorts the stage groups by descending order
stage_groups = stage_groups.sort_values(ascending=False)
stage_groups
```

```python
# define axis values
x_axis = list(stage_groups)
y_axis = [i.title() for i in stage_groups.index]

# plot the bar chart
plt.figure(figsize=[10,5])
plt.xlim(8,13)
plt.xlabel('Average Rating')
plt.title('Average Rating for each Dog Stage on WeRateDogs™')
colors = ['#8dafbe' if (x < max(x_axis)) else '#f8e863' for x in x_axis]
ax = sns.barplot(x=x_axis, y=y_axis, label=y_axis, color="b", palette=colors)
plt.show();
```
